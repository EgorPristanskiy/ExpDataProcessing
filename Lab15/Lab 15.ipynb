```python
import numpy as np
from matplotlib import pyplot as plt
import numpy.random
from numpy.linalg import inv
import random
import pylab
%matplotlib inline
```

```python
ρ = np.loadtxt('ro.txt')
ϕ = np.loadtxt('fi.txt')

Im1 = np.loadtxt('Image1.txt')
Im2 = np.loadtxt('Image2.txt')
Im3 = np.loadtxt('Image3.txt')
Im4 = np.loadtxt('Image4.txt')
Im = np.array((Im1, Im2, Im3, Im4))

Bord1 = np.loadtxt('Borders1.txt')
Bord2 = np.loadtxt('Borders2.txt')
Bord3 = np.loadtxt('Borders3.txt')
Bord4 = np.loadtxt('Borders4.txt')
```

```python
def sort(Bord, ρ, ϕ):
    ρ1 = np.zeros((256,256))
    ϕ1 = np.zeros((256,256))
    k = 0
    buf1 = []
    buf2 = []
    for i in range(256):
        for j in range(256):
             if ρ[i,j] > Bord[0] and ρ[i,j] < Bord[1]:
                    ρ1[i,j] = ρ[i,j]
                    ϕ1[i,j] = ϕ[i,j]
                    buf1.append(ρ1[i,j])
                    buf2.append(ϕ1[i,j])
    Z1 = np.array((buf2, buf1)).transpose()
    return ρ1, ϕ1, Z1
```

```python
fig = plt.figure(figsize = (10,20), dpi=300)
Image = ['Image 1', 'Image 2', 'Image 3', 'Image 4']
for i in range(4):
    
    ax1 = fig.add_subplot(4, 2, i+1)
    ax1 = ax1.tick_params(bottom=False, left=False, labelbottom=False, labelleft=False)
    plt.title(Image[i], size = 20)
    plt.imshow(Im[i], cmap = 'jet')
    xvals = numpy.arange (0, 250)
    yvals = numpy.sinc (xvals)
    pylab.plot (xvals, yvals)
    pylab.text (125, 25, u'N', family="serif", size = 'x-large')
    pylab.text (20, 125, u'E', family="serif", size = 'x-large')
    pylab.text (225, 125, u'W', family="serif", size = 'x-large')
    pylab.text (125, 240, u'S', family="serif", size = 'x-large')

plt.show()
```

```python
ρ1, ϕ1, Z1 = sort(Bord1, ρ, ϕ)
ρ2, ϕ2, Z2 = sort(Bord2, ρ, ϕ)
ρ3, ϕ3, Z3 = sort(Bord3, ρ, ϕ)
ρ4, ϕ4, Z4 = sort(Bord4, ρ, ϕ)
ρ_buf = np.array((ρ1, ρ2, ρ3, ρ4))

Title2 =['Borders 1', 'Borders 2', 'Borders 3', 'Borders 4']

fig = plt.figure(figsize = (10,20), dpi=300)
#plt.title('Front of EUV wave', size = 'xx-large')
for i in range(4):
    ax1 = fig.add_subplot(4, 2, i+1)
    ax1 = ax1.tick_params(bottom=False, left=False, labelbottom=False, labelleft=False)
    plt.imshow(ρ_buf[i], cmap = 'jet')
    xvals = numpy.arange (0, 250)
    yvals = numpy.sinc (xvals)
    pylab.plot (xvals, yvals)
    pylab.text (90, 240, Title2[i], color = 'w', family="serif", size = 'x-large')

plt.show()
```
