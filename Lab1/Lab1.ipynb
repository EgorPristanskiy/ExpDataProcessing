```python
year, month, solar_flux, sunspot_number = np.loadtxt("data_group6.txt", unpack=True)
```

```python
plt.plot(year + month / 12, solar_flux, 'b')
plt.plot(year + month / 12, sunspot_number, 'r')
plt.show()
```

```python
mask = np.ones(13) / 12
mask[0] = 1/24
mask[12] = 1/24
smooth_sunspot_number = sunspot_number.copy()
smooth_solar_flux = solar_flux.copy()
smooth_sunspot_number[6:-6] = np.convolve(sunspot_number[6:-6], mask, mode='same')
smooth_solar_flux[6:-6] = np.convolve(solar_flux[6:-6], mask, mode='same')
plt.plot(smooth_solar_flux, 'b')
plt.plot(smooth_sunspot_number, 'r')
plt.show()
print(smooth_sunspot_number[:6])
print(sunspot_number[:6])
```
